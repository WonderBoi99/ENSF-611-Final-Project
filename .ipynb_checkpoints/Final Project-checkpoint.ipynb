```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import cross_validate


def get_regressor_r2_score(model, X, y, cv=7):
    '''Calculate train and validation score for regressor (model) using cross-validation
        
        score is r2 score
        
        model (sklearn classifier): Regressor to train and evaluate
        X (numpy.array or pandas.DataFrame): Feature matrix
        y (numpy.array or pandas.Series): Target vector
        
        returns: mean training r2 score, mean validation r2 score
    
    '''
    results = cross_validate(model, X, y, cv = cv, scoring = 'r2', return_train_score = True)
    train_scores = results['train_score']
    test_scores = results['test_score']
    train_scores_mean = sum(train_scores) / len(train_scores)
    test_scores_mean =  sum(test_scores) / len(test_scores)
    return train_scores_mean, test_scores_mean
```

# Dataset Details

```python
from yellowbrick.datasets.loaders import load_bikeshare
data = load_bikeshare(return_dataset = True)
print(data.README)
```

```python
# checking size of dataset
X, y = load_bikeshare()
print("Size of X: ", X.shape)
print("Size of y: ", y.shape)
```

```python
# checking max and min values of dataset features
print('Min values in each feature: ')
print(X.min())
print('Max values in each feature: ')
print(X.max())
```

```python
sns.boxplot(data = X)
plt.ylabel('units')
plt.title('Boxplot Of The Features')
plt.show()
```

```python
results = X.corr()
sns.heatmap(results, vmin=-1, vmax=1, annot=True, cmap='BrBG')
plt.title('Correlation Heatmap Of Features')
plt.show()
```

```python
sns.displot(y)
plt.title('Histrogram Of Target Variable')
plt.show()
```

# Start of ML

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=37, test_size=0.2)
```

# Testing Models with their default parameters

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
models = [DecisionTreeRegressor(random_state=22), RandomForestRegressor(random_state=64), GradientBoostingRegressor(random_state=79)]
for model in models:
    train_score, test_score = get_regressor_r2_score(model, x_train, y_train, 5)
    print(model,": ")
    print('Training_score =  %.2f' %train_score)
    print('Testing_score =  %.2f' %test_score)
```

Brief Reflection:<br>
Decision Trees - training score is perfect, this make sense because default parameters of max_depth and max_leaf_nodes is set to None. This means that the depth will go until pure leaf are found (bascially perfection) and it can unlimited leaf nodes.<br>
RandomForests - default parameter for n-estimator is 100 and depths is None. This means that the number of tress in the forest are 100 and depth will go until pure leafs are found. Therefore, makes sense why training score is almost perfect.<br>
GraidentBoosting - default parameter for learning rate is 0.1 and default for max_features is None. Learning rate specify how much each tree contributes and is limited to 0.1 by default. This might be the main reason why the training score are not as good when compared to decision and randomforests trees.<br>

# Modifying Parameters for Decision Trees

```python
depths = [5, 10, 15]
leafs = [30, 85, 100]
for d in depths:
    for l in leafs:
        model = DecisionTreeRegressor(max_leaf_nodes=l, max_depth=d, random_state=22)
        train_score, test_score = get_regressor_r2_score(model, x_train, y_train, 5)
        print('max_depth =', d, ', max_leaf_nodes =', l,": ")
        print('Training_score =  %.2f' %train_score)
        print('Testing_score =  %.2f' %test_score)
```

Brief Reflection:
Best Model is when max_depth is 10/15 and when max_leaf_nodes is 100. This makes sense that the best model includes the the max choice of the max_leaf_nodes. I notice that max_leaf_nodes and max_depth affect each other. For example, if we limit the max_depth to like 2 then the max_leaf_nodes we can have is only 4. 

# Modifying Parameters for Random Forests

```python
depths = [5, 10, 15]
n_estimators = [10, 30, 50]
for d in depths:
    for e in n_estimators:
        model = RandomForestRegressor(n_estimators=e, max_depth=d, random_state=64)
        train_score, test_score = get_regressor_r2_score(model, x_train, y_train, 5)
        print('max_depth =', d, ', n_estimators =', e,": ")
        print('Training_score =  %.2f' %train_score)
        print('Testing_score =  %.2f' %test_score)
```

Brief Reflection:
When the max_depth was 5 and 10 I noticed that no matter what the n_estimators changed to the score stayed the same. I only started seeing n_estimators affect the scores when the max_depth was set to 15. The best model was when max_depth is 15 and n_estimators is 30/50

# Modifying Parameters for Gradient Boosting

```python
rates = [0.2, 0.4, 0.6]
num_of_features = [2, 6, 9]
for r in rates:
    for n in num_of_features:
        model = GradientBoostingRegressor(learning_rate=r, max_features=n , random_state=79)
        train_score, test_score = get_regressor_r2_score(model, x_train, y_train, 5)
        print('learning_rate =', r, ', num_of_features =', n,": ")
        print('Training_score =  %.2f' %train_score)
        print('Testing_score =  %.2f' %test_score)
```

Brief Reflection:
It seems like increasing the learning rate and num_of_features both always have a positive impact on the model and impove the scores. The best model is when learning rate is 0.6 and num_of_features is 9.

# More Analysis with Best Decision Tree model

```python
model = DecisionTreeRegressor(max_leaf_nodes=100, max_depth=10, random_state=22)
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y_train_result = model.predict(x_train)
y_test_result = model.predict(x_test)
print('Mean-squared error for training set = %.2f ' %mean_squared_error(y_train_result, y_train, squared=False))
print('Mean-squared error for testing set = %.2f ' %mean_squared_error(y_test_result, y_test, squared=False))
print('R-squared for training set = %.2f' %r2_score(y_train_result, y_train))
print('R-squared for testing set = %.2f' %r2_score(y_test_result, y_test))
```

```python
sns.scatterplot(x = y_test_result, y = y_test)
sns.lineplot(x = [0, y_test_result.max()], y = [0, y_test_result.max()], color = "orange")
plt.ylabel('actual number of riders')
plt.xlabel('predicted number of riders')
plt.title('Predicted vs Actual number of riders')
```

```python
difference = y_test - y_test_result
sns.scatterplot(x = y_test_result, y = difference)
plt.ylabel('difference between actual and predicted number of riders')
plt.xlabel('predicted number of riders')
plt.title('Residual Plot')
```
